```python
#Age
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]


#Fare
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

```python
sns.boxplot(x='Age', data =df)
plt.title('Outlier Handling in Age column')
plt.show()


sns.boxplot(x='Fare', data =df)
plt.title('Outlier Handling in Fare column')
plt.show()
```

#Interview Questions:


1. What are the different types of missing data?
There are three main types:


MCAR (Missing Completely at Random) – This is the best-case scenario. Data is missing for no reason at all—totally random.

MAR (Missing at Random) – The missing data has something to do with other known variables. For example, younger passengers might be less likely to report their income.

MNAR (Missing Not at Random) – The reason the data is missing is tied to the value itself. Like someone not reporting income because it’s really high or really low.



2. How do you handle categorical variables?
We need to convert categories (like "male", "female", "first class") into numbers, since machine learning models can't read text. Two common ways:

One-hot encoding – This creates new columns for each category (e.g., male, female) and uses 1s and 0s. It's great when the categories don’t have any order.

Label encoding – This just gives each category a number (like 0, 1, 2). It’s best for things that do have an order, like small, medium, large.

3. What's the difference between normalization and standardization?
Both scale your numbers, but in slightly different ways:

Normalization squishes all values between 0 and 1. It’s handy when features are on very different scales.

Standardization centers the data around 0 and gives it a standard deviation of 1. This is better for most models like SVM, logistic regression, etc.

4. How do you detect outliers?
Outliers are data points that are way off compared to the rest. You can find them by:

Plotting boxplots (those show you visually what's way too high or too low)

Using stats like IQR (interquartile range) — if something is way beyond the normal range, it’s probably an outlier

Z-scores can also help. If a value is more than 3 standard deviations away from the mean, it’s suspicious

5. Why is preprocessing important in ML?
Think of preprocessing as cleaning up your kitchen before cooking. If the ingredients are messy or mislabeled, the final dish (your model) won’t turn out well.

Good preprocessing:

Cleans messy data

Fills in blanks

Puts things on the same scale

Makes sure your model isn’t learning from noise or biased data

6. One-hot vs Label Encoding — what's the difference?
One-hot encoding makes a new column for each category and puts 1 or 0. It’s super clear and works great when there’s no natural order (like colors or cities).

Label encoding just gives each category a number. It’s shorter and simpler, but can trick some models into thinking the categories have a ranking (like 0 < 1 < 2), which isn’t always true.



7. How do you handle data imbalance?
When one class (like survivors vs non-survivors) is way more common, the model might get lazy and just always guess the majority class.

To fix that, you can:

Oversample the minority class (add more of the underrepresented data)

Undersample the majority class (remove some of the overrepresented data)

Or tell the model to give more weight to the minority class (most ML libraries support this)

Also, instead of just using accuracy, look at things like F1-score or ROC-AUC, which give you a better picture.

8. Can preprocessing affect model accuracy?

A Big yes!

If your data is messy or unscaled, your model can make bad predictions or learn the wrong patterns. But if you clean it up properly—handle missing values, scale things well, and encode categories—you’ll usually see a big jump in performance.
